```python
# Create a simple 10x10 image filled with zeros (black)
image = np.zeros((10, 10))

# Print the array to see what an image looks like as numbers
print(image)
```

```python
# Create a 10x10 image filled with zeros
image = np.zeros((10, 10))

# Add some bright pixels manually
image[2, 5] = 1.0    # row 2, column 5
image[5, 5] = 0.5    # row 5, column 5
image[8, 5] = 0.25   # row 8, column 5

# Display the image
plt.imshow(image, cmap='gray')
plt.colorbar()
plt.title('My first image')
plt.show()
```

```python
# Create a 100x100 image with random noise
# This simulates the background noise in a real fluorescence microscopy image
image_noise = np.random.normal(loc=0.2, scale=0.05, size=(100, 100))

# Add a bright circular signal in the middle
# This simulates a fluorescently labeled cell
y, x = np.ogrid[0:100, 0:100]
center_y, center_x = 50, 50
radius = 15
circle = (x - center_x)**2 + (y - center_y)**2 <= radius**2
image_noise[circle] = 0.8

# Display it
plt.figure(figsize=(6, 6))
plt.imshow(image_noise, cmap='gray')
plt.colorbar()
plt.title('Simulated fluorescence image')
plt.show()
```
